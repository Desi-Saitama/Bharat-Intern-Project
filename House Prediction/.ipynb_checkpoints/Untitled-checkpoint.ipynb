```python
def drop_columns(X):
    droped = ['id', 'Postal Code', 'Distance from the airport', 'Number of schools nearby', 'lot_area_renov', 'Built Year']
    return X.drop(droped, axis =1)
```

```python
from pandas.plotting import scatter_matrix

attributes = ["number of bedrooms", "living area", "Area of the house(excluding basement)", "Price"]

scatter_matrix(housing[attributes], figsize=(12,8));
```

```python
housing = drop_columns(housing)
```

```python
X_label = housing['Price']
X = housing.drop(['Price'], axis = 1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

feed_data = Pipeline([
    ('std_scaler', StandardScaler()),
    ])
```

```python
housing_tr = feed_data.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_tr, X_label)
```

```python
some_data = X.iloc[:5]
some_labels = X_label.iloc[:5]
some_data_prepared = feed_data.transform(some_data)
```

```python
print('Predictions:', lin_reg.predict(some_data_prepared))
```

```python
print("Labels: ", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error

housing_prediction = lin_reg.predict(housing_tr)
lin_mse = mean_squared_error(X_label, housing_prediction)
lin_mse = np.sqrt(lin_mse)
lin_mse
```

```python
from sklearn.metrics import mean_absolute_error as mae

lin_mae = mae(X_label, housing_prediction)
lin_mae
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lin_reg, X, X_label,
                         scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(lin_rmse_scores)
```
